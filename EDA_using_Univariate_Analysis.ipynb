```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('train.csv')
```

```python
df.sample(5)
```

# **1. Categorial Data**
## a. countplot


```python
# With Matplot Survived or Not at titanic
df['Survived'].value_counts().plot(kind='bar')
```

```python
# With seaborn
sns.countplot(x='Survived',data=df)
plt.show()
```

```python
# With matplot
df['Pclass'].value_counts().plot(kind='bar')
```

```python
# With Seaborn Passenger Class
sns.countplot(x='Pclass',data=df)
plt.show()
```

## b. Piechart


```python
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')  # autopct is used for label at pie chart
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```

#  2. Numerical Data
## a. Histogram

```python
# Histogram for Age

plt.hist(df['Age'])
```

## b. Distplot

```python
sns.displot(data=df, x='Age', kde=True)
```
