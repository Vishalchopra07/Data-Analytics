```python
# program for Analysis for total sales state wise , product wise and gender wise
import matplotlib.pyplot as plt

customer=["Vishal","Pulkit","Khushi","Harsh","Vibha","Ashu","Vanshika","Arun","Vikas","Vansh"]
product=["Fridge","AC","Computer","Fridge","Laptop","Computer","Laptop","AC","Computer","Laptop"]
purchase=[15000,35000,15000,25000,50000,30000,65000,38000,42000,37000]
amount=[20000,40000,20000,30000,60000,40000,70000,45000,50000,45000]
state=["Haryana","Punjab","Himachal","Haryana","Rajasthan","Haryana","Himachal","Punjab","Himachal","Rajasthan"]
gender=["Male","Male","Female","Male","Female","Male","Female","Male","Male","Male"]
print("Customer   :",customer)
print("")
print("Product    :",product)
print("")
print("Amount     :",amount)
print("")
print("State      :",state)
print("")
print("Gender     :",gender)
print("")

#Finding profit
profits=[]
for profit in range(len(amount)):
 profits.append(amount[profit]-purchase[profit])
print("Profit     :",profits)
print("")

#Finding Total sales

total_sales=0
for sales in amount:
  total_sales=total_sales + sales
print("Total Sales:",total_sales,"Rupees")
print("")

#Finding state wise sales

print("Now Find State Wise Sales")
print("")
unique_state=[]
state_wise_sale=[]
for states in state:
  if states not in unique_state:
    unique_state.append(states)
    state_wise_sale.append(0)
print("Unique State",unique_state)
for i in range(len(state)):
  index=unique_state.index(state[i])
  state_wise_sale[index]=state_wise_sale[index]+amount[i]
print("State Wise Sales",state_wise_sale)
print("")

#Finding state wise profit
state_wise_profit=[]
for s in unique_state:
 state_wise_profit.append(0)
print("Unique State :",unique_state)
for a in range(len(state)):
  index3=unique_state.index(state[a])
  state_wise_profit[index3]=state_wise_profit[index3]+profits[a]
print("State Wise Profit :",state_wise_profit)

# Plotting graph of sales
colors=["blue","red","purple","pink"]
plt.bar(unique_state,state_wise_sale,color=colors)
plt.xlabel("State")
plt.ylabel("sales")
plt.show()
print("")
print("Now Find Product Wise Sales")
print("")
unique_product=[]
product_wise_sales=[]

#Finding product wise sales

for products in product:
 if products not in unique_product:
  unique_product.append(products)
  product_wise_sales.append(0)
print("Unique Product :",unique_product)

for j in range(len(product)):
  index1=unique_product.index(product[j])
  product_wise_sales[index1]=product_wise_sales[index1]+amount[j]
print("")
print("Product Wise Sales: ",product_wise_sales,"Rupees")
print("")

#plotting graph of product wise sales
plt.bar(unique_product,product_wise_sales)
plt.xlabel("Product")
plt.ylabel("Product Wise Sales")
plt.show()

# Now finding gender wise sales

unique_gender=[]
gender_wise_sales=[]
for genders in gender:
  if genders not in unique_gender:
    unique_gender.append(genders)
    gender_wise_sales.append(0)
print("Unique Gender: ",unique_gender)

for k in range(len(gender)):
  index2=unique_gender.index(gender[k])
  gender_wise_sales[index2]=gender_wise_sales[index2]+amount[k]
print("Gender Wise Sales :",gender_wise_sales,"Rupees")

plt.bar(unique_gender,gender_wise_sales,color="red")
plt.xlabel("Gender")
plt.ylabel("Gender Wise Sales")
plt.show
```
