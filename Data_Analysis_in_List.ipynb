```python
import matplotlib.pyplot as plt

customer=["Vishal","Pulkit","Khushi","Harsh","Vibha","Ashu","Vanshika","Arun","Vikas","Vansh"]
product=["fridge","AC","Cooler","Fridge","Laptop","Computer","Laptop","AC","Computer","Cooler"]
amount=[20000,40000,20000,30000,60000,40000,70000,45000,50000,45000]
state=["Haryana","Punjab","Himachal","Haryana","Rajasthan","Haryana","Himachal","Punjab","Himachal","Rajasthan"]
gender=["Male","Male","Female","Male","Female","Male","Female","Male","Male","Male"]
print("Customer   :",customer)
print("Product    :",product)
print("Amount     :",amount)
print("State      :",state)
print("Gender     :",gender)
total_sales=0
for i in amount:
  total_sales=total_sales + i
print("Total Sales:",total_sales,"Rupees")
unique_state=[]
state_wise_sale=[]
for j in state:
  if j not in unique_state:
    unique_state.append(j)
    state_wise_sale.append(0)
print("Unique State",unique_state)
for k in range(len(state)):
  index=unique_state.index(state[k])
  state_wise_sale[index]=state_wise_sale[index]+amount[k]
print("State Wise Sales",state_wise_sale)

plt.bar(unique_state,state_wise_sale)
plt.xlabel("State")
plt.ylabel("sales")
plt.show()

```
