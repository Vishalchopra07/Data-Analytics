# **gender wise Score**
## using matplot


```python
# gender wise math score

x=(df.groupby('gender')['math score'].sum().plot(kind='bar'))
plt.show(x)
```

```python
# gender wise reading score

y=(df.groupby('gender')['reading score'].sum().plot(kind='bar'))
plt.show(y)
```

```python
# gender wise writing score
z=(df.groupby('gender')['writing score'].sum().plot(kind='bar'))
plt.show(z)
```

```python
# student's level of education

df['parental level of education'].value_counts().plot(kind='pie',autopct='%.2f')
```
